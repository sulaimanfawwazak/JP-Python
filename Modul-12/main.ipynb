### **Mengeskpresikan Fungsi dengan Deret Taylor**

#### **01: JALANKAN! Gunakan Python untuk memplot fungsi sin bersama dengan perkiraan deret Taylor orde pertama, ketiga, kelima, dan ketujuh**

````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize=(10, 8))
for n, label in zip(range(4), labels):
  y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
  plt.plot(x, y, label=label)

plt.plot(x, np.sin(x), 'k', label='Analytic')
plt.grid()
plt.title('Taylor Series Approximation of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

## **Diferensiasi Numerik**

### **01: JALANKAN! Perhatikan fungsi f(x)=cos(x). Kita tahu turunan dari cos(x) adalah sin(x). Meskipun dalam praktiknya kita mungkin tidak mengetahui fungsi dasar yang kita temukan turunannya, kita menggunakan contoh sederhana untuk mengilustrasikan metode diferensiasi numerik yang disebutkan di atas dan keakuratannya. Kode berikut menghitung turunan secara numerik.**

````python
# Step size
h = 0.1
# Define grid
x = np.arange(0, 2*np.pi, h)
# Compute function
y = np.cos(x)

# Compute vector of forward differences
forward_diff = np.diff(y) / h
# Compute corresponding grid
x_diff = x[:-1:]
# Compute exact solution
exact_solution = -np.sin(x_diff)

# Plot solution
plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Finite difference approximation')
plt.plot(x_diff, exact_solution, label='Exact solution')
plt.legend()
plt.show()

# Compute max error between numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(f'Max error: {max_error}')
````

### **02: JALANKAN! Kode berikut menghitung turunan numerik dari f(x)=cos(x) menggunakan rumus selisih maju untuk mengurangi ukuran langkah, h. Ini kemudian memplot kesalahan maksimum antara turunan yang didekati dan turunan sebenarnya versus h seperti yang ditunjukkan pada gambar yang dihasilkan.**

````python
# Define step size
h = 1
# Define number of iterations to perform
iterations = 20
# List to store our step sizes
step_size = []
# List to store max error for each step size
max_error = []

for i in range(iterations):
  # Halve the step size
  h = h/2
  # Store this step size
  step_size.append(h)
  # Compute new grid
  x = np.arange(0, 2 * np.pi, h)
  # Compute function value at grid
  y = np.cos(x)
  # Compute vector of forward differences
  forward_diff = np.diff(y) / h
  # Compute corresponding grid
  x_diff = x[:-1:]
  # Compute exact solution
  exact_solution = -np.sin(x_diff)

  # Compute max error between
  # Numerical derivative and exact solution
  max_error.append(max(abs(exact_solution - forward_diff)))

# Produce log-log plot of max error versus step size
plt.figure(figsize=(12, 8))
plt.loglog(step_size, max_error, 'v')
plt.show()
````

## **Integrasi Numerik**

### **Integral Riemann**

#### **01: JALANKAN! Gunakan Integral Riemann kiri, Integral Riemann kanan, dan Aturan Mid Point untuk mengaproksimasi fungsi berikut Bandingkan nilai ini dengan nilai yang tepat.**

````python
a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)

x = np.linspace(a, b, n)
f = np.sin(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - I_riemannL

I_riemannR = h * sum(f[1::])
err_riemannR = 2 - I_riemannR

I_mid = h * sum(np.sin((x[:n-1] + x[1:])/2))
err_mid = 2 - I_mid

print(I_riemannL)
print(err_riemannL)
print(I_riemannR)
print(err_riemannR)
print(I_mid)
print(err_mid)
````

### **Aturan Trapesium**

#### **01: JALANKAN! Gunakan Aturan Trapesium untuk mengaproksimasi fungsi berikut. Bandingkan nilai ini dengan nilai yang eksak**

````python
a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)

x = np.linspace(a, b, n)
f = np.sin(x)

I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])
err_trap = 2 - I_trap

print(I_trap)
print(err_trap)
````

### **Aturan Simpson**

#### **01: JALANKAN! Gunakan Aturan Simpson untuk mengaproksimasi fungsi berikut.**

````python
a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)

x = np.linspace(a, b, n)
f = np.sin(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp

print(I_simp)
print(err_simp)
````

### **Menghitung Integral dengan Python**

### **01: JALANKAN! Gunakan fungsi trapz untuk mengaproksimasi fungsi berikut. Bandingkan nilai ini dengan yang dihitung pada contoh awal menggunakan Aturan Trapesium.**

````python
from scipy.integrate import trapz

a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)

x = np.linspace(a, b, n)
f = np.sin(x)

I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print(I_trapz)
print(I_trap)
````

#### **02: JALANKAN! Gunakan fungsi cumtrapz untuk mengaproksimasi integral kumulatif f(x)=sin(x) dari 0 ke  dengan langkah diskritisasi 0,01. Solusi eksak dari integral ini adalah F(x)=sin(x). Plot hasilnya.**

````python
from scipy.integrate import cumtrapz
%matplotlib inline

x = np.arange(0, np.pi, 0.01)
F_exact = -np.cos(x)
F_approx = cumtrapz(np.sin(x), x)

plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} sin(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
````
